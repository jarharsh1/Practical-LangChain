```python
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
```

```python
from langchain_core.documents import Document

doc1 = Document(
    page_content="Royal Enfield Classic 350 is one of the most loved bikes in India for its retro design and comfortable riding posture. It is powered by a 349 cc engine and is known for smooth cruising and strong road presence. It suits riders who enjoy relaxed city rides and long highway journeys.",
    metadata={
        "bike" : "Royal Enfield Classic 350"
    }
)

doc2 = Document(
    page_content="Bajaj Pulsar NS200 is a popular sporty motorcycle among young riders. It comes with a 199.5 cc liquid-cooled engine and offers strong performance at an affordable price. The bike is known for its aggressive styling, good handling, and punchy acceleration.",
    metadata={
        "bike" : "Bajaj Pulsar NS200"
    }
)

doc3 = Document(
    page_content="TVS Apache RTR 160 4V is a well-balanced bike that combines performance with everyday usability. It uses a 159.7 cc engine and delivers good mileage along with sharp handling. It is widely chosen for daily commuting as well as occasional spirited riding.",
    metadata={
        "bike" : "TVS Apache RTR 160 4V"
    }
)

doc4 = Document(
    page_content="Hero Splendor Plus is India’s most trusted commuter bike. It is powered by a 97.2 cc engine and is famous for excellent mileage, low maintenance cost, and long-term reliability. It is ideal for daily use and is widely used across cities and rural areas.",
    metadata={
        "bike" : "Hero Splendor Plus"
    }
)

doc5 = Document(
    page_content="KTM Duke 390 is a high-performance street motorcycle designed for enthusiasts. It features a 373 cc engine and offers advanced features, strong power delivery, and sharp design. It is best suited for riders who want performance-oriented city and highway riding.",
    metadata={
        "bike" : "KTM Duke 390"
    }
)
```

```python
docs = [doc1,doc2,doc3,doc4,doc5]
```

```python
embeddings = OpenAIEmbeddings(
    api_key="**************************",
    model="text-embedding-3-small"
)

vector_store = Chroma(
    embedding_function = embeddings,
    persist_directory = "chroma_db",
    collection_name = "bikes"
)
```

```python
vector_store.add_documents(docs)
```

```python
vector_store.get(include=['embeddings','documents','metadatas'])
```

```python
vector_store.similarity_search(
    query= "which bike is best for village driving",
    k=2
)
```

```python
vector_store.similarity_search_with_score(
    query= "which bike is best for village driving",
    k=2
)
```

```python
vector_store.similarity_search_with_score(
    query= "",
    filter={"bike":'TVS Apache RTR 160 4V'}
)
```
